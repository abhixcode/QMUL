```python
#Q1

import numpy as np
np.set_printoptions(precision=5)

xTrain = np.array([1,2,3,4])
yTrain = np.array([8,8,10,10])

#Liner Model
order = 1
w = np.polyfit(xTrain, yTrain, order)
p = np.poly1d(w)
print("\n The coefficients of the model are : \n ", p.c)
```

```python
a = np.array([2,3,4,5,6])
b = np.array([1,1,1,1,1])

mx = np.mean(a)
my = np.mean(b)

print("\nMean of arr : ", ( mx ) )
print("Mean of arr2 : ", ( my ) )

## Euclidean distance for cluster 1
euclidean = 0
for i in range(len(a)):
    v1 = np.array((a[i],b[i]))
    for j in range(i,len(a)):         
        v2 = np.array((a[j],b[j]))
        euclidean += (np.sum((v1 - v2) ** 2)) #  sqrt removed here
        print(v1, ' - ', v2)
print("Squared Distance C4 = " + str((euclidean)))
```
